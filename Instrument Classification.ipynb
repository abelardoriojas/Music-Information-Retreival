```python
#we know how to make mlps now, let's see if we can use the MFCCs of each sample in a 3 second waveform of an 
#instrument to classify them.

import numpy as np
from random import random
from sklearn.model_selection import train_test_split
import tensorflow as tf


#instruments are in folder: IRMAS-TrainingData
```

```python
#making the data
import os
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
folder_name = 'IRMAS-TrainingData'
sub_folders = [
'cel',
'cla',
'voi',
'gac',
'gel',
'org',
'pia',
'sax',
'tru',
'vio',
'flu']


audio_mfccs = []
targets = []
itera = 0
for sub in sub_folders:
    directory = folder_name + '/' + sub
    for filename in os.listdir(directory):
        print('filename: {}'.format(filename))
        audio_file = directory + '/' + filename
        signal, sr = librosa.load(audio_file)
        mfccs = librosa.feature.mfcc(signal, n_mfcc=20, sr=sr)
        audio_mfccs.append(mfccs)#get those mfccs
        target = np.zeros(11)#length of subfoldiers
        for i,sub in enumerate(sub_folders):
            if sub in filename:
                target[i] = 1
        targets.append(target)
        print(target)
        print(itera/6705)
        itera+=1
audio_mfccs = np.array(audio_mfccs)
targets = np.array(targets)
        
                
                
                
        
        
        
        

```

```python
audio_mfccs.shape
#6705 waveforms, 20 mfccs, 130 frames

x = []
for waveform in audio_mfccs:
    x.append(waveform.flatten())
    
x = np.array(x)

x.shape
    
```

```python
targets.shape
```

```python
# create a dataset with 6705 samples
x_train, x_test, y_train, y_test = train_test_split(x, targets, test_size=.3)


# build model with 4 layers: 2600 --> 1305 --> 24 --> 11
model = tf.keras.models.Sequential([ #sequential=inputs move left to right
    tf.keras.layers.Dense(1305, input_dim=2600, activation="sigmoid"), #dense because all neurons are connected
    tf.keras.layers.Dense(24, activation="sigmoid"), #hidden  layer
    tf.keras.layers.Dense(11, activation="sigmoid") #output layer
])

# choose optimiser
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)#stochastic gradient descent

# compile model
model.compile(optimizer=optimizer, loss='mse')#mean squared error

# train model
model.fit(x_train, y_train, epochs=1000, shuffle=True)

# evaluate model on test set
print("\nEvaluation on the test set:")
model.evaluate(x_test,  y_test, verbose=2)
```

```python
#evaluate

audio_file = 'IRMAS-TrainingData/gel/[gel]shesaid.wav'



signal, sr = librosa.load(audio_file)
signal_3s = signal[66150*4:66150*5]
mfcc = librosa.feature.mfcc(signal_3s, n_mfcc=20, sr=sr)

mfcc = np.array([mfcc.flatten()])
print(mfcc.shape)



pred = list(model.predict(mfcc)[0])

print(pred)
maxval = max(pred)

index = pred.index(maxval)

print(sub_folders[index])



```
